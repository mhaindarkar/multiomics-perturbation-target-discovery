## Rank score rule

For each perturbation:

- up genes get descending positive scores
- down genes get ascending negative scores

Example:
- first up gene gets the highest positive score
- first down gene gets the most negative score

## 5. Helper function to convert up/down lists into ranked signatures


```python
def create_signature(perturbation_id, perturbation_name, up_genes, down_genes):
    df_up = pd.DataFrame({
        "perturbation_id": perturbation_id,
        "perturbation_name": perturbation_name,
        "gene_symbol": [g.strip().upper() for g in up_genes],
        "rank_score": list(range(len(up_genes), 0, -1))
    })

    df_down = pd.DataFrame({
        "perturbation_id": perturbation_id,
        "perturbation_name": perturbation_name,
        "gene_symbol": [g.strip().upper() for g in down_genes],
        "rank_score": list(range(-len(down_genes), 0))
    })

    df = pd.concat([df_up, df_down], ignore_index=True)

    df["abs_rank"] = df["rank_score"].abs()
    df = df.sort_values("abs_rank", ascending=False)
    df = df.drop_duplicates(subset=["perturbation_id", "gene_symbol"], keep="first")
    df = df.drop(columns=["abs_rank"])
    df = df.sort_values(["perturbation_id", "rank_score"], ascending=[True, False]).reset_index(drop=True)

    return df
```

## 6. Build the combined ranked signature table


```python
all_signatures = pd.concat([
    create_signature("drug_001", "dexamethasone", dex_up, dex_down),
    create_signature("drug_002", "budesonide", bud_up, bud_down),
    create_signature("drug_004", "tofacitinib", tof_up, tof_down),
    create_signature("drug_005", "ruxolitinib", rux_up, rux_down),
], ignore_index=True)

all_signatures.head(20)
```

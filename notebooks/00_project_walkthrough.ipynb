## 1. Problem Definition

Many diseases are driven by coordinated pathway dysregulation rather than changes in single genes alone.

The goal of this project is to identify therapeutic perturbations and candidate targets by:

1. detecting disease-associated gene expression changes,
2. converting these changes into pathway-level disease signatures,
3. identifying perturbations that reverse these disease pathways,
4. validating candidate targets using single-cell and proteomics evidence,
5. integrating all evidence into a final perturbation and target ranking.


## 2. Core Idea

Disease → DEG → GSEA  
Perturbation → GSEA → Anticorrelation  
Single-cell → Cell-type context  
Proteomics → Orthogonal validation  
Integration → Final prioritization

A strong candidate should:

- be associated with disease-relevant dysregulation,
- participate in disease-relevant pathways,
- be linked to perturbations that reverse the disease signature,
- be supported by disease-relevant cell types,
- and ideally show support at protein level.


## 3. Disease Focus

Version 1 of this repository focuses on **ulcerative colitis (UC)**.

Why UC is a good first disease model:

- strong inflammatory biology,
- public bulk transcriptomics datasets are available,
- compatible single-cell datasets exist,
- proteomics validation can be layered in,
- pathway-level interpretation is biologically meaningful.

### Main disease dataset
- **GSE235236**
- bulk transcriptomics
- main contrast used in the project: **UC vs HC**


## 4. Data Layers Used

This project integrates multiple biological data layers.

### Disease transcriptomics
- bulk RNA-seq / count matrix
- used to define the disease signature

### Perturbation data
- curated LINCS-style perturbation signatures
- used to identify pathway-reversing perturbations

### Single-cell RNA-seq
- used to identify which cell types support candidate targets
- implemented with a real Scanpy workflow

### Proteomics
- used as an orthogonal validation layer for target plausibility


## 5. Pipeline Overview

The project is built in the following order:

1. QC and disease expression table generation  
2. Disease pathway enrichment analysis (GSEA)  
3. Perturbation pathway profiling  
4. Anticorrelation analysis  
5. Single-cell validation  
6. Proteomics validation  
7. Multi-omics integration  
8. Final summary reporting

## 8. Final Biological Questions

The project is designed to answer two related questions:

> Which perturbations most strongly reverse disease-associated pathway activity?

and

> Which target genes are most strongly supported by integrated multi-omics evidence?

The final output is therefore not just a gene list.

It is a biologically reasoned framework for:
- perturbation prioritization
- target plausibility assessment
- multi-layer evidence integration


## 9. What Success Looks Like

A successful final repository should clearly answer:

1. What is the disease-associated molecular signal?  
2. Which pathways are dysregulated?  
3. Which perturbations reverse the disease pathway state?  
4. Which cell types support the candidate targets?  
5. Is the result supported at protein level?  
6. Which perturbations and targets are most convincing after integration?

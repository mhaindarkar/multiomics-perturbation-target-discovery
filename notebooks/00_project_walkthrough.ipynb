## 1. Problem Definition

Many diseases are driven by coordinated pathway dysregulation rather than changes in single genes alone.

The goal of this project is to identify therapeutic targets by:

1. detecting disease-associated gene expression changes,
2. converting these changes into pathway-level disease signatures,
3. finding perturbations that reverse these disease pathways,
4. validating candidate targets using single-cell and proteomics evidence,
5. integrating all evidence into a final target ranking.


## 2. Core Idea

Disease -> DEG -> GSEA  
Perturbation -> GSEA -> Anticorrelation  
Single-cell -> Cell-type origin  
Proteomics -> Validation  
Integration -> Final target prioritization

A good target should:

- be dysregulated in disease,
- belong to disease-relevant pathways,
- be reversed by perturbation signatures,
- be supported by disease-driving cell types,
- and ideally show support at protein level.


## 3. Disease Focus

Version 1 of this repository focuses on **ulcerative colitis (UC)**.

Why UC is a good first disease model:

- strong inflammatory biology,
- public bulk transcriptomics datasets are available,
- compatible single-cell datasets exist,
- proteomics validation datasets are available,
- pathway-level interpretation is biologically meaningful.


## 4. Data Layers Used

This project integrates multiple biological data layers:

### Transcriptomics
- bulk RNA-seq or microarray
- used to define the disease signature

### Perturbation data
- CRISPR knockout or drug perturbation signatures
- used to identify reversal candidates

### Single-cell RNA-seq
- used to identify which cell types drive the disease signal

### Proteomics
- used to test whether transcript-level changes are also visible at protein level


## 5. Pipeline Overview

The project is built in the following order:

1. Quality control and differential expression analysis  
2. Disease pathway enrichment analysis (GSEA)  
3. Perturbation pathway profiling  
4. Anticorrelation analysis  
5. Single-cell validation  
6. Proteomics validation  
7. Multi-omics integration  
8. Final target prioritization
